```r
library(Stat2Data)
```

```r
data(SeaIce)
str(SeaIce)
```

```r
plot(SeaIce$Extent ~ SeaIce$t, xlab = 'Time since Year 1(1979)', ylab = 'Extent of SeaIce', main = 'Extent of SeaIce from 1979')
```

This scatterplot fits the negative relationship. The value of extent are decrease when the year passing. Also, the relationship is relatively strong. 
The regression line will be a little bit curve.

```r
model_extent = lm(Extent ~ t, data = SeaIce)
plot(model_extent$residuals ~ model_extent$fitted.values, xlab = 'Predicted value of SeaIce', ylab = 'Residual value', main = 'Residual plot')
```

It has obvious curvilinear relationship and cannot be modeled by linear model.

```r
model_square_it = SeaIce$Extent^2
plot(model_square_it ~ SeaIce$t, xlab = 'Time since Year 1(1979)', ylab = 'Extent of SeaIce', main = 'Extent of SeaIce from 1979')
```

It still a little bit curved, but better than previous one. Can be try modeling using linear models.

```r
model_squared = lm(model_square_it ~ t, data = SeaIce)
plot(model_squared$residuals ~ model_squared$fitted.values, xlab = 'Predicted value of SeaIce', ylab = 'Residual value', main = 'Residual plot')
```

This model is better than non-square model.
