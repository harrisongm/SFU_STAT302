In the data of 2003, Fall is 343 and Spring is 288. This is a big concern point.

```r
springAndFall = lm(Spring ~ Fall, data = MathEnrollment)
summary(springAndFall)
```

```r
newDataFall = c(259, 301, 307, 286, 273, 248, 292, 250, 278, 303)
newDataSpring = c(246, 206, 215, 230, 247, 308, 271, 285, 286, 254)
newModel = lm(newDataSpring ~ newDataFall)
summary(newModel)
```

Because we delete the data from 2003, the new least square line's slope has changed a lot, from -0.3266 to -1.0483. So, this point can be a influential point.
