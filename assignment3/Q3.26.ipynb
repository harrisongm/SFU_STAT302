```r
### Q3.26
install.packages('Stat2Data', repos='http://cran.us.r-project.org')
library(Stat2Data)
data(Tadpoles)
str(Tadpoles)
```

```r
#part a
plot(x=Tadpoles$Body, y=Tadpoles$GutLength, xlab="Body", ylab="GutLength", main="Scatterplot", 
     col=Tadpoles$Treatment)
legend("topleft", legend = levels(Tadpoles$Treatment), pch = 16, col = factor(levels(Tadpoles$Treatment)))
```

For both Bd and control treatments, body and Gutlength have positive, linear relationships.

```r
#part b
lmModel = lm(GutLength ~ Body, data=Tadpoles)
summary(lmModel)
```

t = 4.204, df = 25, P-value = 0.000293, reject H0, has linear relationship.

```r
#part c
prl_model = lm(GutLength ~ Body + Treatment, data=Tadpoles)
summary(prl_model)
```

For Bd:
GutLength-hat = 15.47 + 8.07Body
For Control:
GutLength-hat = 15.47 + 8.07Body + 16.28*1 = 31.75 + 8.07Body

```r
#part d
new_fitted_model = lm(GutLength ~ Body + Treatment + MouthpartDamage, data=Tadpoles)
summary(new_fitted_model)
```

Yes, the treatment and Gutlength still have a positive relationship given MouthpartDamage constant.
