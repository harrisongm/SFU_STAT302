```r
### Q3.24
install.packages('Stat2Data', repos='http://cran.us.r-project.org')
library(Stat2Data)
data(LewyBody2Groups)
str(LewyBody2Groups)
```

```r
#part a
plot(x=LewyBody2Groups$APC,y=LewyBody2Groups$MMSE,col=LewyBody2Groups$Type,
     xlab="APC", ylab="MMSE", main="Functional performance in Examination")
legend("topleft", legend = levels(LewyBody2Groups$Type), pch = 16, 
       col = factor(levels(LewyBody2Groups$Type)))
```

For different types, DLB, DLB/AD, APC and MMSE have POSITIVE, LINEAR relationships.

```r
#part b
model = lm(MMSE~APC, data=LewyBody2Groups)
summary(model)
```

t = 3.968, df = 37, P-value = 0.00321, rejected H0. We have evidence of a linear relationship between APC and MMSE

```r
#part c
model1 = lm(MMSE~APC+Type, data=LewyBody2Groups)
summary(model1)
```

For type DLB/AD:
MMSE-hat = -0.9433 + 1.5015APC - 1.3135*1 = -2.2568 + 1.5015APC

For type DLB:
MMSE-hat = -0.9433 + 1.5015APC
